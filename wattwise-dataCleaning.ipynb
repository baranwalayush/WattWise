1. Extract only Delhi's demand data from original datasets
2. Combine differnet years data in a single dataframe
3. Create a datetime columns by extracting the year and datetime details from the State and Date columns
4. Drop the unimportant columns
5. Get the csv file of the cleaned data

```python
data = [delhi2024, delhi2023, delhi2022, delhi2021, delhi2020, delhi2019, delhi2018, delhi2017]
```

```python
df = pd.concat(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df['State'] = df['State'].astype(str)
df['year'] = df['State'].str.extract(r'(\d{4})').astype(int)
```

```python
df['datetime_str'] = df['Date'] + ' ' + df['year'].astype(str)
```

```python
df['datetime'] = pd.to_datetime(df['datetime_str'], format='%d-%b %I%p %Y', errors='coerce')
```

```python
df.head()
```

```python
df = df.drop(columns=['Date', 'year', 'datetime_str'])
```
